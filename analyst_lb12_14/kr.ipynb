# Визуализация.

Датасет представляет собой коллекцию данных об академических достижениях студентов, их профессиональных навыках и карьерном успехе после окончания университета. Данные включают информацию об образовании, опыте, soft skills, нетворкинге, а также о начальной зарплате и карьерном росте выпускников. Этот набор данных может быть использован для анализа факторов, влияющих на успех в карьере, прогнозирования уровня зарплаты и определения ключевых компетенций, необходимых для успешного старта в профессиональной жизни.

В будущем этот датасет может быть применен для построения моделей машинного обучения, которые помогут предсказать уровень зарплаты или вероятность трудоустройства на основе академической успеваемости, количества стажировок, проектов и других факторов. Это позволит студентам лучше планировать свою карьеру, а работодателям — находить наиболее перспективных кандидатов.

Описание столбцов:

|Название| Описание |
|--|--|
|`Student_ID`| Уникальный идентификатор студента (не несет полезной информации для анализа). |
|`Age`| Возраст варьируется от 18 до 29 лет. |
|`Gender`| Пол студента. Может принимать значения "Male", "Female" или "Other". |
|`High_School_GPA`| Средний балл успеваемости в школе. Диапазон значений от 2.0 до 4.0. |
|`SAT_Score`| Результаты теста SAT. Диапазон значений от 900 до 1600. | 
|`University_Ranking`| Рейтинг университета, который закончил студент. Диапазон от 1 до 1000. |
|`University_GPA`| Средний балл успеваемости в университете. Диапазон значений от 2.0 до 4.0. | 
|`Field_of_Study`| Направление обучения (например, инженерия, экономика, информатика и т.д.). |
|`Internships_Completed`| Количество завершенных стажировок. Диапазон от 0 до 4. |
|`Projects_Completed`| Количество завершенных проектов. Диапазон от 0 до 9. |
|`Certifications`| Количество полученных сертификатов. Диапазон от 0 до 5. |
|`Soft_Skills_Score`| Оценка soft skills (мягких навыков) студента. Диапазон от 1 до 10. |
|`Networking_Score`| Оценка навыков нетворкинга студента. Диапазон от 1 до 10. |
|`Job_Offers`| Количество предложений работы после окончания университета. Диапазон от 0 до 5. |
|`Starting_Salary`| Начальная зарплата после трудоустройства. Диапазон от 25,000 до 101,000. |
|`Career_Satisfaction`| Уровень удовлетворенности карьерой. Диапазон от 1 до 10. |
|`Years_to_Promotion`| Количество лет до первого повышения по службе. Диапазон от 1 до 5. |
|`Current_Job_Level`| Текущий уровень работы (например, junior, middle, senior). |
|`Work_Life_Balance`| Оценка баланса между работой и личной жизнью. Диапазон от 1 до 10. |
|`Entrepreneurship`| Информация о предпринимательской деятельности. Может принимать значения "Yes" или "No". |

## Визуализация

```python
numeric = df.select_dtypes(include=['float64', 'int64']).columns.tolist()
categorical = df.select_dtypes(include=['object']).columns.tolist()
```

```python
# Гистограммы для каждого числового признака
for col in numeric:
    plt.figure(figsize=(8, 5))
    sns.histplot(df[col], bins=30, kde=True, color="skyblue", edgecolor="black")
    plt.title(f"Гистограмма {col}", fontsize=14)
    plt.xlabel(col, fontsize=12)
    plt.ylabel("Частота", fontsize=12)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
```

```python
# Ящики с усами для каждого числового признака
for col in numeric:
    plt.figure(figsize=(8, 5))
    sns.boxplot(x=df[col], color="lightgreen")
    plt.title(f"Ящик с усами для {col}", fontsize=14)
    plt.xlabel(col, fontsize=12)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
```

```python
# Анализ категориальных переменных
for col in categorical:
    plt.figure(figsize=(8, 5))
    sns.countplot(x=col, data=df, palette="Set2", order=df[col].value_counts().index, hue=col, legend=False)
    plt.title(f"Распределение {col}", fontsize=14)
    plt.xlabel(col, fontsize=12)
    plt.ylabel("Количество", fontsize=12)
    plt.xticks(rotation=45, fontsize=10)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
```

```python
# Матрица корреляции для числовых признаков
plt.figure(figsize=(12, 8))
sns.heatmap(df[numeric].corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5, square=True)
plt.title("Матрица корреляции (Пирсон)", fontsize=16)
plt.show()
```

Это указывает на то, что студенты начинают учебу в этом возрасте (около 18 лет), а также есть группа старших студентов, которые могут быть на повторном обучении или продолжают образование после работы. Наблюдаются выбросы для тех, кто завершил 4 стажировки, что может указывать на высокую мотивацию и стремление получить больше практических навыков. Частота завершенных проектов имеет более равномерное распределение по сравнению со стажировками. Большинство студентов завершают от 2 до 6 проектов, при этом пик частоты находится на значении 3-4 проекта. Это подтверждает важность практического опыта в рамках учебного процесса. ## Сравнение выборок

```python
# Зависимость зарплаты от GPA в школе
plt.figure(figsize=(10, 5))
sns.scatterplot(x='High_School_GPA', y='Starting_Salary', data=df, alpha=0.6, color="orange")
plt.title("Зависимость зарплаты от GPA в школе", fontsize=14)
plt.xlabel("High_School_GPA", fontsize=12)
plt.ylabel("Starting_Salary", fontsize=12)
plt.grid(axis='both', linestyle='--', alpha=0.7)
plt.show()

# Зависимость зарплаты от GPA в университете
plt.figure(figsize=(10, 5))
sns.scatterplot(x='University_GPA', y='Starting_Salary', data=df, alpha=0.6, color="purple")
plt.title("Зависимость зарплаты от GPA в университете", fontsize=14)
plt.xlabel("University_GPA", fontsize=12)
plt.ylabel("Starting_Salary", fontsize=12)
plt.grid(axis='both', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Зависимость зарплаты от SAT_Score
plt.figure(figsize=(10, 5))
sns.scatterplot(x='SAT_Score', y='Starting_Salary', data=df, alpha=0.6, color="green")
plt.title("Зависимость зарплаты от SAT Score", fontsize=14)
plt.xlabel("SAT_Score", fontsize=12)
plt.ylabel("Starting_Salary", fontsize=12)
plt.grid(axis='both', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Группировка по Internships_Completed и Projects_Completed
grouped = df.groupby(['Internships_Completed', 'Projects_Completed'])['Starting_Salary'].mean().reset_index()
pivot = grouped.pivot(index='Internships_Completed', columns='Projects_Completed', values='Starting_Salary')

# Тепловая карта зависимости зарплаты от стажировок и проектов
plt.figure(figsize=(10, 8))
sns.heatmap(pivot, annot=True, cmap="YlGnBu", fmt=".0f", linewidths=0.5)
plt.title("Зависимость зарплаты от стажировок и проектов", fontsize=14)
plt.xlabel("Projects_Completed", fontsize=12)
plt.ylabel("Internships_Completed", fontsize=12)
plt.show()
```

```python
# Зависимость зарплаты от Networking_Score
plt.figure(figsize=(10, 5))
sns.scatterplot(x='Networking_Score', y='Starting_Salary', data=df, alpha=0.6, color="red")
plt.title("Зависимость зарплаты от Networking Score", fontsize=14)
plt.xlabel("Networking_Score", fontsize=12)
plt.ylabel("Starting_Salary", fontsize=12)
plt.grid(axis='both', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Зависимость зарплаты от Soft_Skills_Score
plt.figure(figsize=(10, 5))
sns.scatterplot(x='Soft_Skills_Score', y='Starting_Salary', data=df, alpha=0.6, color="blue")
plt.title("Зависимость зарплаты от Soft Skills Score", fontsize=14)
plt.xlabel("Soft_Skills_Score", fontsize=12)
plt.ylabel("Starting_Salary", fontsize=12)
plt.grid(axis='both', linestyle='--', alpha=0.7)
plt.show()
```

**Вывод:**
<p>Студенты с высокой зарплатой в среднем старше, чем студенты с низкой зарплатой. Это может быть связано с тем, что более старшие студенты чаще имеют больше опыта или выбирают специальности, требующие дополнительной подготовки. Высокий GPA как в школе, так и в университете коррелирует с более высокой начальной зарплатой. Студенты с высоким результатом SAT чаще получают высокооплачиваемые должности. Студенты с высокой зарплатой в среднем завершили больше стажировок и проектов. Практический опыт явно является ключевым фактором успеха.</p>
<p>Поскольку графики Networking_Score и Soft_Skills_Score показывают похожую зависимость от зарплаты, можно сделать вывод, что оба параметра имеют схожую роль в определении уровня зарплаты. Высокие оценки по Networking_Score и Soft_Skills_Score (7–8) чаще встречаются у студентов с высокой зарплатой. Однако общая корреляция со стартовой зарплатой для этих параметров остается умеренной, хотя они важны для успешного поиска работы и могут повлиять на количество предложений.</p>
<p>Как следствие вышеупомянутых факторов, студенты с высоким уровнем образования, опытом и навыками получают более высокую начальную зарплату 60,000–100,000 против 25,000–50,000.</p>
<p>Распределение мужчин и женщин примерно одинаково в обеих группах, что указывает на отсутствие гендерного влияния на уровень зарплаты в данном датасете. Студенты с высокой зарплатой чаще выбирают технические специальности (инженерия, информатика, финансы), тогда как студенты с низкой зарплатой чаще выбирают гуманитарные направления или менее востребованные специальности. Большинство студентов начинают карьеру как junior специалисты. Небольшая часть студентов с высокой зарплатой также занимается предпринимательской деятельностью.</p>
<p>Высокий GPA и результаты SAT имеют положительную корреляцию с начальной зарплатой, но не являются единственными решающими факторами. Опыт и навыки играют более важную роль.</p>
<p>Количество завершенных стажировок и проектов является одним из ключевых факторов, влияющих на уровень зарплаты. Студенты с большим практическим опытом чаще занимают высокооплачиваемые должности.</p>
<p>Софт скилс и нетворкинг имеют умеренную корреляцию с зарплатой, но они важны для успешного поиска работы и могут повлиять на количество предложений. В частности, высокие оценки по Networking_Score и Soft_Skills_Score часто встречаются у студентов с высокой зарплатой, что подчеркивает их значимость в профессиональном развитии.</p>
